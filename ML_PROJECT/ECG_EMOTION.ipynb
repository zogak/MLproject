```python
diff['DIFF_Joy'] = diff.apply(lambda x : x['POST_Joy'] - x['PRE_Joy'], axis=1)
diff['DIFF_Calmness'] = diff.apply(lambda x : x['POST_Calmness'] - x['PRE_Calmness'], axis=1)
diff['DIFF_Relaxation'] = diff.apply(lambda x : x['POST_Relaxation'] - x['PRE_Relaxation'], axis=1)
diff['DIFF_Happy'] = diff.apply(lambda x : x['POST_Happy'] - x['PRE_Happiness'], axis=1)

diff['DIFF_Anger'] = diff.apply(lambda x : x['POST_Anger'] - x['PRE_Anger'], axis=1)
diff['DIFF_Sadness'] = diff.apply(lambda x : x['POST_Sadness'] - x['PRE_Sadness'], axis=1)
diff['DIFF_Disgust'] = diff.apply(lambda x : x['POST_Disgust'] - x['PRE_Disgust'], axis=1)
diff['DIFF_Fear'] = diff.apply(lambda x : x['POST_Fear'] - x['PRE_Fear'], axis=1)
diff['DIFF_Anxiousness'] = diff.apply(lambda x : x['POST_Anxiousness'] - x['PRE_Anxiousness'], axis=1)

diff['DIFF_Arousal'] = diff.apply(lambda x : x['POST_Arousal'] - x['PRE_Arousal'], axis=1)
diff['DIFF_Valence'] = diff.apply(lambda x : x['POST_Valence'] - x['PRE_Valence'], axis=1)
```

```python
diff_data = diff.iloc[:,[0,1,2,3,4,5,30,31,32,33,34,35,36,37,38]]
```

```python
diff_data
```

Quad_Cat

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
X = diff_data.iloc[:, 6:]
y = diff_data.loc[:, ['Quad_Cat']]

train_X, test_X, train_y, test_y = train_test_split(X,y, train_size=0.8, test_size=0.2, random_state = 42)
```

```python
model = GaussianNB()
model.fit(train_X, train_y)
```

```python
pred_test = model.predict(test_X)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_y, pred_test)
```
